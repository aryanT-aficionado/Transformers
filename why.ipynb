```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create spotlight effect
x = np.linspace(-3, 3, 400)
y = np.exp(-x**2)  # Gaussian curve = spotlight beam

plt.figure(figsize=(8, 3), facecolor='black')
plt.plot(x, y, color='#FF6F00', lw=8, alpha=0.9)  # Warm orange beam
plt.fill_between(x, y, color='#FF6F00', alpha=0.3)

# Add text in the spotlight
plt.text(0, 0.5, "ATTENTION", fontsize=30, fontweight='bold',
         color='white', ha='center', va='center',
         bbox=dict(boxstyle="round,pad=0.4", facecolor='#FF6F00', alpha=0.9))

# Clean up axes
plt.axis('off')
plt.title("Why Transformers? â†’ They Pay ATTENTION.", 
          color='white', fontsize=16, pad=20, fontweight='bold')
plt.tight_layout()
plt.savefig("transformer_spotlight.png", dpi=150, facecolor='black')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sentence tokens
tokens = ["The", "bank", "by", "the", "river", "had", "financial", "issues"]

# Simulated attention weights for the word "bank" (focusing on "river" for context)
attention_weights = np.array([0.1, 1.0, 0.3, 0.2, 0.9, 0.1, 0.4, 0.2])  # Manual weights for demo

# Create a heatmap
plt.figure(figsize=(8, 3))
sns.heatmap(
    [attention_weights],  # Reshape to 2D for heatmap
    xticklabels=tokens,
    yticklabels=["Attention for 'bank'"],
    cmap="YlOrRd",
    annot=True,
    fmt=".2f",
    cbar=False
)
plt.title("How Transformers Understand Context: Attention on 'bank'");
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
